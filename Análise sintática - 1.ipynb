```python
"""
Gramática:

Σ = {0,1,2,3, x, y, z, begin, end, input, :=, print, (,), +,-,*,/}

conjuntos auxiliares : literal = {0,1,2,3,}

                                : id = {x,y,z}

P →begin L end
L →C L | ε
C →[id := E | input(id) | print(id)];
E →T R
R →[+|−] T R| ε
T →F S
S →[∗|/]F S | ε
F →(E) | id | literal

begin
x:=3
print(x)
input(x)
end
"""
```

```python
class LL1:
    def __init__(self,input):
        self.input = input
        self.vazio = [' ','\n','\t']
        self.id = ['x','y','z']
        self.literal = ['0','1','2','3']
    
    def proximo(self):
        return self.input[0]
    
    def avanca(self):
        self.input = self.input[1:]
    
    def nda(self):
        while self.proximo() in self.vazio:
            self.avanca()
    
    def p(self):
        print("P -> begin L end")
        self.begin()
        self.l()
        self.end()
        
    def begin(self):
        palavra = ''
        self.nda()
        while self.proximo() in ['b','e','g','i','n']:
            palavra += self.proximo()
            self.avanca()
        if palavra == 'begin':
            pass
        else:
            raise Exception('Erro no begin')
        del palavra
        
    def l(self):
        print('L -> C L | ε')
        self.c()
        self.l()
        
    def end(self):
        self.nda()
        palavra = ''
        print(self.proximo())
        while self.proximo() in ['e','n','d']:
            palavra +=self.proximo()
            self.avanca()
        print("palavra = " + palavra)
        if palavra == 'end':
            self.avanca()
        else:
            raise Exception('Erro no end')
            
    def c(self):
        self.nda()
        rule = ''
        if self.proximo() in self.id:
            self.avanca()
            self.nda()
            if self.proximo() == ':':
                self.avanca()
                if self.proximo() == '=':
                    self.avanca()
                    rule = 'id'
                    self.e()
                else:
                    raise Exception('Erro no id')
            else:
                raise Exception('Erro no id')
        else:
            palavra = ''
            while self.proximo() in ['i','n','p','u','t','(',')','r']+self.id:
                palavra = self.proximo()
                self.avanca()
            possiveis_inputs = ['input(x)','input(y)','input(z)','print(x)','print(y)','print(z)']
            if palavra in possiveis_inputs:
                return
            else:            
                raise Exception('Erro no id')
            
    def e(self):
        print("E -> T R")
        self.t()
        self.avanca()
        self.r()
        
    def t(self):
        print("T-> F S")
        self.f()
        self.s()
    
    def s(self):
        self.nda()
        print(self.proximo())
        rule = ''
        if self.proximo() == '*':
            rule = '*'
        elif self.proximo() == '/':
            rule = '/'
        print(rule + ' F S | ε')
        self.f()
        self.s()
        
    def f(self):
        rule = ''
        if self.proximo() == "(":
            rule = self.proximo()
            self.avanca()
            self.e()
            self.avanca() 
            if self.proximo == ")":
                rule+=')'
            else:
                raise Exception("Erro no )")
        elif (self.proximo() in self.id):
            rule = self.proximo()
            self.avanca()
        elif self.proximo() in self.literal:
            rule = self.proximo()
            self.avanca()
        print('F-> '+ rule)
            
        
    def parse(self):
        try:
            self.p()
            if self.proximo() == '$':
                return True
        except:
            return False
        return False
```
